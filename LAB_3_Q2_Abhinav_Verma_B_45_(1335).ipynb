```python
!pip install librosa matplotlib

import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the audio file from URL
import requests

audio_url = "https://github.com/Jakobovski/free-spoken-digit-dataset/raw/master/recordings/0_george_0.wav"
audio_path = "/tmp/audio.wav" # Define a local path to save the audio file

# Download the audio file
r = requests.get(audio_url)
with open(audio_path, 'wb') as f:
    f.write(r.content)

y, sr = librosa.load(audio_path, sr=None)  # y=audio array, sr=sampling rate

print(f"Original max value: {np.max(y)}")
print(f"Original min value: {np.min(y)}")
```

```python
# Step 2: Normalize the audio signal (amplitude between -1 and 1)
y_norm = y / np.max(np.abs(y))
print(f"Normalized max value: {np.max(y_norm)}")
print(f"Normalized min value: {np.min(y_norm)}")
```

```python
# Step 3: Visualize the audio waveform
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y_norm, sr=sr)
plt.title("Normalized Audio Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Step 4: Extract MFCC features
mfccs = librosa.feature.mfcc(y=y_norm, sr=sr, n_mfcc=13)  # 13 is standard for MFCCs
print("MFCCs shape:", mfccs.shape)
```

```python
# Visualize MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', sr=sr)
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
